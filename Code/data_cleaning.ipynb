```python
import pandas as pd
```

```python
# Read CSV file
df = pd.read_csv('E:\\CS-6400-Database\\Data\\assigned.csv')
```

```python
df.head()
```

```python
# Split the 'CategoryName' by commas and explode the DataFrame
df = df.assign(districts_assigned=df['districts_assigned'].str.split(',')).explode('districts_assigned')
```

```python
# Remove leading/trailing whitespace from 'CategoryName'
df['districts_assigned'] = df['districts_assigned'].str.strip()
```

```python
# Display the reformatted DataFrame
print(df.head(10))
```

```python
df.to_csv('E:\\CS-6400-Database\\Demo Data\\reformatted_assigned.csv', index=False)
```

```python
df.shape
```
